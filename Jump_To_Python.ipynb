```
##커피머신 최종판!!!##
coffee=10
while True:
  money= int(input('돈을 넣어주세요'))
  if money==300:
    print('커피가 나옵니다')
    coffee-=1
    print(f"커피가 {coffee}개 남았습니다")
  elif money>300:
    print('커피가 나옵니다')
    coffee-=1
    print(f"잔돈:{money-300}원")
    print(f"커피가 {coffee}개 남았습니다")
  else:
    while True:                                                          # 무한루프 안의 무한 루프!
      money=money+int(input(f"{300-money}원을 더 넣으세요:"))            # money변수 갱신!!
      if money>300:
         print('커피가 나옵니다')
         coffee-=1
         print(f"잔돈:{money-300}원")
         print(f"커피가 {coffee}개 남았습니다")
         break
      elif money==300:
        print('커피가 나옵니다')
        coffee-=1
        print(f"커피가 {coffee}개 남았습니다")
        break
  if coffee==0:
    print('품절')
    break
```
