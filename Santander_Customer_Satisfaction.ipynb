```python
# 이상치 제거 함수 만들기
import numpy as np

def get_outlier_1(df=None, column=None, weight=1.5):
  satisfied = df[df['TARGET']==1][column]
  quantile_25= np.percentile(satisfied.values,25)
  quantile_75= np.percentile(satisfied.values,75)

  iqr=quantile_75 - quantile_25
  iqr_weight=iqr*weight
  lowest_val=quantile_25 - iqr_weight
  highest_val=quantile_75 + iqr_weight

  outlier_index=satisfied[(satisfied<lowest_val)|(satisfied>highest_val)].index.tolist()

  return outlier_index
```

```python
import numpy as np

def get_outlier_0(df=None, column=None, weight=1.5):
  satisfied = df[df['TARGET']==0][column]
  quantile_25= np.percentile(satisfied.values,25)
  quantile_75= np.percentile(satisfied.values,75)

  iqr=quantile_75 - quantile_25
  iqr_weight=iqr*weight
  lowest_val=quantile_25 - iqr_weight
  highest_val=quantile_75 + iqr_weight

  outlier_index=satisfied[(satisfied<lowest_val)|(satisfied>highest_val)].index.tolist()

  return outlier_index
```

```python
def remove_outlier1(df=None, cols=None, weight=1.5):
  list1=[]
  if len(cols)>1:
    for col in cols:
      outlier_index_1=get_outlier_1(df=df, column=col, weight=weight)
      list1.extend(outlier_index_1)   
  else:
    col="".join(cols)
    outlier_index_1=get_outlier_1(df=df, column=col, weight=weight)
    list1.extend(outlier_index_1)

  a=set(list1)
  df2=df.drop(a, axis=0)
  return df2
```

```python
def remove_outlier0(df=None, cols=None, weight=1.5):
  list1=[]
  if len(cols)>1:
    for col in cols:
      outlier_index_0=get_outlier_0(df=df, column=col, weight=weight)
      list1.extend(outlier_index_0)   
  else:
    col="".join(cols)
    outlier_index_0=get_outlier_0(df=df, column=col, weight=weight)
    list1.extend(outlier_index_0)

  a=set(list1)
  df2=df.drop(a, axis=0)
  return df2
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score

def lgb_roc_eval(learning_rate,num_leaves,colsample_bytree,subsample,max_depth,reg_alpha,\
                 reg_lambda,min_split_gain,min_child_weight):
  
  params={'n_estimators':200,'learning_rate':learning_rate,'num_leaves':\
          int(round(num_leaves)),'colsample_bytree':colsample_bytree,'subsample':\
          subsample,'max_depth':int(round(max_depth)),'reg_alpha':reg_alpha,\
          'reg_lambda':reg_lambda,'min_split_gain':min_split_gain,'min_child_weight':\
          min_child_weight,'verbosity':-1}

  lgb_model=LGBMClassifier(**params)
  lgb_model.fit(X_train,y_train,eval_set=[(X_test,y_test)],early_stopping_rounds=30,\
                eval_metric='auc',verbose=100)
  valid_proba=lgb_model.predict_proba(X_test)[:,1]
  roc_preds=roc_auc_score(y_test,valid_proba)
  return roc_preds
```
